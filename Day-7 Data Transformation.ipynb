## Satandard Deviation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([67,54,17,83,14,59,63,77])
```

```python
plt.scatter(x,range(8))
```

```python
plt.scatter(x,range(8))
plt.scatter(x.mean(),3.5,color='red')
```

```python
std =np.sqrt(np.sum(( x-x.mean())**2)/len(x))
```

```python
std
```

```python
x.std()
```

## Standardization

```python
# It make standard dev. 1 for the dataset
def standardize(x):
    val=(x-x.mean())/x.std()
    return  val
```

```python
x_new=standardize(x)
```

```python
x_new.std()
```

```python
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.scatter(x,range(8))

plt.subplot(1,2,2)
plt.title('Standardize Data')
plt.scatter(x_new,range(8))
```

## Data Resampling/Balancing

```python
df['Purchased'].value_counts()
```

```python
plt.pie(df['Purchased'].value_counts(),labels=df['Purchased'].value_counts().index,autopct='%2.2f%%');
```

## Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus=RandomUnderSampler(random_state=0)
```

```python
y=df['Purchased']
```

```python
x_new,y_new=rus.fit_resample(x,y)
```

```python
y_new.value_counts()
```

## Oversampler

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=0)
x_new,y_new=ros.fit_resample(x,y)
```

```python
y_new.value_counts()
```

## Feature Selection

```python
df=pd.read_csv('wisc_bc_data.csv')
df
```

```python
# Separate input and output variable

x=df.drop(['id','diagnosis'],axis=1) #feature
y=df['diagnosis']   #outcome
```

```python
x.shape
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skf=SelectKBest(score_func=chi2,k=5)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x_new = x.iloc[:,skf.get_support()]
x_new
```
